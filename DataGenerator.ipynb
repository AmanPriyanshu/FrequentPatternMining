```python
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
import pickle
```

```python
data = pd.read_csv("./FrequentPatternMining/data/u.data", header=None, sep='\t')
data.columns = ["User-Id", "Movie-Id", "Rating", "Timestamp"]
data.head()
```

```python
movies = pd.read_csv("./FrequentPatternMining/data/u.item", header=None, sep="|", encoding='latin-1')
movies.columns = ['movie id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movies.head()
```

```python
def generate_ui_matrix(data, movies):
  data_np = data.values
  movies_np = movies.values
  ui_matrix = np.zeros((np.max(data_np.T[0]), np.max(data_np.T[1])))
  for row in tqdm(data_np):
    ui_matrix[row[0]-np.min(data_np.T[0])][row[1]-np.min(data_np.T[1])] = row[2]
  ui_matrix_df = pd.DataFrame(ui_matrix, columns = [movie_name for movie_name in movies_np.T[1]], index=["User"+str(i+1) for i in range(ui_matrix.shape[0])])
  return ui_matrix, ui_matrix_df
```

```python
_, ui_matrix = generate_ui_matrix(data, movies)
ui_matrix.head()
```

```python
ui_matrix.to_csv("ui_matrix.csv")
```

```python
def user_compressed_lists_generator(ui_matrix_df):
  movie_names = list(ui_matrix_df.columns)
  usernames = list(ui_matrix_df.index)
  ui_matrix = ui_matrix_df.values
  compressed_lists = []
  for row in tqdm(ui_matrix):
    indexes = np.argwhere(row>=1).flatten()
    compressed_lists.append({"Movie-Names": [movie_names[i] for i in indexes], "indexes": [i for i in indexes]})
  return compressed_lists
```

```python
user_compressed_lists = user_compressed_lists_generator(ui_matrix)
```

```python
with open('user_compressed_lists.pkl', 'wb') as f:
  pickle.dump(user_compressed_lists, f)
```
